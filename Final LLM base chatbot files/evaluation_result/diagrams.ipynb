```python
import numpy as np
import matplotlib.pyplot as plt

# Define Sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Generate values
x = np.linspace(-10, 10, 400)

# Plot Sigmoid
fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(x, sigmoid(x), label='Sigmoid', color='b', linestyle='-', linewidth=2)
ax.set_title('Sigmoid Activation Function', fontsize=14)
ax.set_xlabel('Input', fontsize=12)
ax.set_ylabel('Output', fontsize=12)
ax.axhline(0, color='black', linewidth=1)
ax.axvline(0, color='black', linewidth=1)
ax.grid(True)
ax.legend(loc='best', fontsize=12)

# Add formula text
formula_text = r'$f(x) = \frac{1}{1 + e^{-x}}$'
ax.text(0.05, 0.9, formula_text, transform=ax.transAxes, fontsize=14, verticalalignment='top')

plt.tight_layout()
plt.savefig("siggmoid.png")
plt.show()
```

```python
# Define ReLU function
def relu(x):
    return np.maximum(0, x)

# Plot ReLU
fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(x, relu(x), label='ReLU', color='g', linestyle='-', linewidth=2)
ax.set_title('ReLU Activation Function', fontsize=14)
ax.set_xlabel('Input', fontsize=12)
ax.set_ylabel('Output', fontsize=12)
ax.axhline(0, color='black', linewidth=1)
ax.axvline(0, color='black', linewidth=1)
ax.grid(True)
ax.legend(loc='best', fontsize=12)

# Add formula text
formula_text_relu = r'$f(x) = \max(0, x)$'
ax.text(0.05, 0.9, formula_text_relu, transform=ax.transAxes, fontsize=14, verticalalignment='top')

plt.tight_layout()
plt.savefig("rellu.png")
plt.show()
```

```python
# Define Tanh function
def tanh(x):
    return np.tanh(x)

# Plot Tanh
fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(x, tanh(x), label='Tanh', color='r', linestyle='-', linewidth=2)
ax.set_title('Tanh Activation Function', fontsize=14)
ax.set_xlabel('Input', fontsize=12)
ax.set_ylabel('Output', fontsize=12)
ax.axhline(0, color='black', linewidth=1)
ax.axvline(0, color='black', linewidth=1)
ax.grid(True)
ax.legend(loc='best', fontsize=12)

# Add formula text
formula_text_tanh = r'$f(x) = \tanh(x)$'
ax.text(0.05, 0.9, formula_text_tanh, transform=ax.transAxes, fontsize=14, verticalalignment='top')

plt.tight_layout()
plt.savefig("tannh.png")
plt.show()
```
