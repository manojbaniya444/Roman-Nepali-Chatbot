```python
# Function to extract model response
def extract_model_response(text):
    match = re.search(r'<start_of_turn>model\n(.*?)<eos>', text, re.DOTALL)
    return match.group(1).strip() if match else None

# Apply function to extract response
df['model_response'] = df['model_generation'].apply(extract_model_response)
```

## EVALUATION METRICS

```python
import evaluate
import nltk
nltk.download('punkt')
```

```python
bleu = evaluate.load("bleu")
```

#### Each Sentence BLEU

```python
def compute_bleu(references, predictions):
    # Tokenize sentences
    tokenized_refs = [[nltk.word_tokenize(ref)] for ref in references]  # Nested for multiple references
    tokenized_preds = [nltk.word_tokenize(pred) for pred in predictions]

    # Compute BLEU
    results = bleu.compute(predictions=tokenized_preds, references=tokenized_refs)
    return results["bleu"]
```

```python
# # Example usage
# df["bleu_score"] = df.apply(lambda row: compute_bleu([row["original_response"]], [row["model_response"]]), axis=1)
```

```python
results = bleu.compute(
    predictions=df["model_response"].tolist(),  # Full sentences, not tokenized
    references=[[ref] for ref in df["response"].tolist()]  # List of lists
)

print("Corpus BLEU Score:", results["bleu"])
```
