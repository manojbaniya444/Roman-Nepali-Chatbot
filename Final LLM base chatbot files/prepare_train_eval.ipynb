```python
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split

dataset = pd.read_csv("./fine_tune_gemma.csv")

train_dataset, eval_dataset = train_test_split(dataset, test_size=2000, random_state=42)
```

```python
train_dataset.shape, eval_dataset.shape
```

```python
train_dataset.head(10)
```

```python
train_category_counts = train_dataset["category"].value_counts()
eval_category_counts = eval_dataset["category"].value_counts()
total_cateogry_counts = dataset["category"].value_counts()

print("Train Dataset Category Counts:")
print(train_category_counts)

print("\nEval Dataset Category Counts:")
print(eval_category_counts)

print("\nTotal Dataset Category Counts:")
print(total_cateogry_counts)
```
