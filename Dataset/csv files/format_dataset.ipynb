## Making dataset

```python
import pandas as pd
```

### New dataset

```python
df_new = pd.read_csv("./after_finetune.csv")
```

```python
df_new.shape
```

```python
df_new.isna().sum()
```

```python
df_new = df_new.fillna(" ")
```

```python
df_new.isna().sum()
```

```python
df_new.sample(5)
```

```python
df_new["category"].unique()
```

```python
def generate_instruction(category):
    if category == 'nep-eng':
        return 'Translate the text below from Nepali to English:'
    elif category == 'eng-nep':
        return 'Translate the text below from English to Nepali:'
    elif category == "RAG":
        return 'Answer the following question by looking at the context if context is given.'
    elif category == "open_qa":
        return 'You are a helpful assistant. Answer the given user question.'
    else:
        return ''

df_new['instruction'] = df_new['category'].apply(generate_instruction)
```

```python
df_new.sample(5)
```

### Old dataset

```python
df_old = pd.read_csv("./v1_fine_tune.csv")
```

```python
df_old.shape
```

```python
df_old.isna().sum()
```

```python
df_old = df_old.fillna("")
```

```python
df_old.sample(5)
```

```python
df_old.drop(columns=["system"], inplace=True)
```

```python
df_old.sample(5)
```

```python
df_old["category"].unique()
```

```python
def generate_instruction_2(category):
    if category == "open_qa":
        return "You are a helpful assistant. Answer the given user question."
    elif category == "RAG":
        return "Answer the following question by looking at the context if context is given."
    elif category == "summary":
        return "Summarize the following text:"
    elif category == "translate":
        return None
    else:
        return ""
```

```python
df_old["instruction"] = df_old.apply(lambda row: generate_instruction_2(row["category"]) if generate_instruction_2(row["category"]) is not None else row["instruction"], axis=1)
```

```python
df_old.sample(10)
```

### Combine the dataset

```python
df_combine = pd.concat([df_old, df_new], ignore_index=True)
```

```python
df_combine.shape
```

```python
df_combine.isna().sum()
```

```python
def format_category(category):
    if category == "eng-nep" or category == "nep-eng":
        return "translate"
    else:
        return category
```

```python
df_combine["category"] = df_combine["category"].apply(format_category)
```

```python
df_combine.sample(20)
```

```python
df_combine_shuffled = df_combine.sample(frac=1).reset_index(drop=True)
df_combine_shuffled.to_csv("v2_fine_tune.csv", index=False)
```

```python
df_combine_shuffled.isna().sum()
```
