## Format Prompt

```python
def format_prompt(row):
    if row["category"] == "RAG":
        prompt = f"""Below is an instruction that describes a task paired with an input that provides further context. Write a response that appriately complete the request.
        
### Instruction
{row["question"]}
        
### Input
{row["context"]}
        
### Response
{row["response"]}"""
        return prompt
    
    elif row["category"] == "open_qa" and row["context"] != " ":
        prompt = f"""Below is an instruction that describes a task paired with an input that provides further context. Write a response that appriately complete the request.

### Instruction
{row["question"]}
        
### Input
{row["context"]}
        
### Response
{row["response"]}"""
        return prompt
    
    elif row["category"] == "open_qa" and row["context"] == " ":
        prompt = f"""Below is an instruction that describes a task. Write a response that appriately complete the request.

### Instruction
{row["question"]}
        
### Response
{row["response"]}"""
        return prompt
    
    elif row["category"] == "translate":
        prompt = f"""Below is an instruction that describes a task paired with an input that provides further context. Write a response that appriately complete the request.

### Instruction
{row["instruction"]}
        
### Input
{row["question"]}
        
### Response
{row["response"]}"""
        return prompt
    
    elif row["category"] == "summary":
        prompt = f"""Below is an instruction that describes a task paired with an input that provides further context. Write a response that appriately complete the request.

### Instruction
{row["instruction"]}
        
### Input
{row["question"]}
        
### Response
{row["response"]}"""
        return prompt
    
    elif row["category"] == "alpaca" and row["context"] != " ":
            prompt = f"""Below is an instruction that describes a task paired with an input that provides further context. Write a response that appriately complete the request.

### Instruction
{row["question"]}
            
### Input
{row["context"]}
            
### Response
{row["response"]}"""
            return prompt
        
    elif row["category"] == "alpaca" and row["context"] == " ":
            prompt = f"""Below is an instruction that describes a task. Write a response that appriately complete the request.

### Instruction
{row["question"]}

### Response
{row["response"]}"""
            return prompt
```

```python
df_combine_shuffled["prompt"] = df_combine_shuffled.apply(format_prompt, axis=1)
```
