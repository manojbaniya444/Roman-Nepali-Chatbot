```python
import json
import pandas as pd
```

```python
file_name = r"D:\\Roman Nepali Chatbot\Dataset\\Instruction Fine Tune Dataset\\data_56.json"
```

```python
df = pd.read_csv("fine_tune.csv")
```

```python
df.tail()
```

```python


with open(file_name, "r") as f:
    data = f.read()
    
json_data = json.loads(data)


new_df = []
for row in json_data:
    question = row["question"]
    context = row["context"]
    response = row["response"]
    system = "Talako instruction herera user le gareko question ko Roman Nepali Language ma answer vana\n\n"
    instruction = "User le gareko question lai, context herera mildo answer deu\n"
    
    new_df.append({
        "category": "RAG",
        "system": system,
        "instruction": instruction,
        "question": question,
        "context": context,
        "response": response
    })

df_new = pd.DataFrame(new_df)
```

```python
df_new.sample(5)
```

```python
df = pd.concat([df, df_new], ignore_index=True)
```

```python
df.tail(5)
```

```python
df.to_csv("fine_tune.csv", index=False)
```
