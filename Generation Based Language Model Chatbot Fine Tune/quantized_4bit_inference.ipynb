```python
from unsloth import FastLanguageModel
import torch
max_seq_length = 2048 # Choose any! We auto support RoPE Scaling internally!
dtype = None # None for auto detection. Float16 for Tesla T4, V100, Bfloat16 for Ampere+
load_in_4bit = True
```

```python
prompt = """
Below is an instruction that describes a task paired with an input that provides further context. Write a response that appriately complete the request.

### instruction:
{instruction}

### question:
{question}

### input:
{context}

### response:
{response}
"""
```

```python
from unsloth import FastLanguageModel
```

```python
model, tokenizer = FastLanguageModel.from_pretrained(
        model_name = "manojbaniya/best_01", # MODEL YOU USED FOR TRAINING
        max_seq_length = max_seq_length,
        dtype = dtype,
        load_in_4bit = load_in_4bit,
    )
```

```python
FastLanguageModel.for_inference(model)
```

```python
from transformers import TextStreamer
text_streamer = TextStreamer(tokenizer)
```

```python
def generate_text(query):
    inputs = tokenizer(
    [
        prompt.format(
            instruction="User le gareko question lai answer deu",
            question=query,
            context="",
            response="", # output - leave this blank for generation!
        )
    ], return_tensors = "pt").to("cuda")
    _ = model.generate(**inputs, streamer = text_streamer, max_new_tokens = 128)
```

```python
generate_text("There are 4 people in this room")
```

```python
generate_text("Nepal ko capital city kaha ho?")
```

```python
generate_text("5 + 5 kati hunxa")
```

```python
generate_text("Mt Everest ko Nepali name ke ho")
```

```python
generate_text("Lionel Messi ko ho?")
```
